```python
#1.记录一周记账功能

income=[]
spending=[]
Balance_every_day={}
Balance=0

#先记录第一天收入支持结余
#在使用for遍历7天收入支出结余最终结余

for day in range(1,8):
    day1_income=int(input("请输入第"+str(day)+"天收入"))
    day1_spending=int(input("请输入第"+str(day)+"天支出"))
    income.append(day1_income)
    spending.append(day1_spending)
    Balance_every_day["第"+str(day)+"天"]=day1_income-day1_spending
    print("")
    print("第"+str(day)+"天收入：",income[day-1],"元")
    print("第"+str(day)+"天支出：",spending[day-1],"元")
    print("第"+str(day)+"天结余：",Balance_every_day["第"+str(day)+"天"],"元")
    print("_______________________________________")
for i in Balance_every_day.values():
        Balance=Balance+i
print("最终结余：",Balance,"元")
```

```python
# 2.ATM机功能:查询余额、存款、取款、退出
"""
1、账户{有则使用，无则新建}
2、查询余额=0
3、存款：余额=余额+存款
4、取款：余额=余额-取款
5、退出

1、循环登录
    2、循环操作，退出后跳出操作循环，进入登录循环·
"""
#state=True

# 2.ATM机功能:查询余额、存款、取款、退出

users=set()
balance={}
    
while True:
    userid=input("请输入姓名：")
    if userid =="关机":
        break
    if userid in users:
        print(userid,"登录成功！")
    else:
        print("正在开卡中doing...\n请稍后...\n恭喜"+userid+"开户成功")
        users.add(userid)
        balance[userid]=0
        
    while True:
        operation=input("请输入操作：存款、取款、查询余额、退出：")
        if operation=="存款":
            deposit=int(input("请输入存款金额："))
            balance[userid]=balance[userid]+deposit
        elif operation=="取款":
            withdrawals=int(input("请输入取款金额："))
            if withdrawals>balance[userid]:
                print("您的余额不足，请重新输入")
            else:
                balance[userid]=balance[userid]-withdrawals
        elif operation=="查询余额":
            print("您当前余额为：",balance[userid])
        elif operation=="退出":
            print("退出账户成功，请收好您的卡片")
            break
        else:
            print("您的输入有误，请重新输入")

###    print(users,balance)
```
